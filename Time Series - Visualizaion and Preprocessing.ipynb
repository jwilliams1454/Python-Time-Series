```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
data = sm.datasets.co2.load_pandas()
co2 = data.data
```

```python
print(co2.head(5))
```

```python
# 1 Indexing
co2.index
```

```python
# Resample weekly to monthly
y = co2['co2'].resample('MS').mean()
```

```python
y.head(5)
```

```python
# only 1990
y['1990':]
```

```python
# between October 1995 and October 1996
y['1995-10-01':'1996-10-01']
```

```python
# 2 Missing values
y.isnull().sum()
```

```python
# fill in missing values with the closest non-null value in our time series
y = y.fillna(y.bfill())
```

```python
# check to see they are filled in
y.isnull().sum()
```

```python
# 3 Visualizing Time-series Data
y.plot(figsize=(15, 6))
plt.show()
```

```python
# 4 Decomposition
# http://www.wessa.net/download/stl.pdf

from pylab import rcParams
rcParams['figure.figsize'] = 11, 9

decomposition = sm.tsa.seasonal_decompose(y, model='additive')
fig = decomposition.plot()
plt.show()
```
